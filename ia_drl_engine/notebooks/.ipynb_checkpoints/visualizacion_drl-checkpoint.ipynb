# Visualización de la IA DRL: Red de Decisiones y Progreso del Estudiante
Este notebook permite:
- Consultar el backend para obtener recomendaciones de la IA.
- Visualizar la red de decisiones (dummy en esta versión).
- Graficar el progreso de un estudiante.

> **Requisitos:**
- Tener el backend corriendo en http://127.0.0.1:8000/
- Tener instalados: requests, networkx, matplotlib, graphviz (puedes instalar con pip).

```python
# Instala dependencias si es necesario (descomenta la línea siguiente)
# !pip install requests networkx matplotlib graphviz
```

```python
import requests
import networkx as nx
import matplotlib.pyplot as plt
from graphviz import Digraph
import random
```

## 1. Consultar el backend para obtener la recomendación de la IA
Simulamos el avance de un estudiante y consultamos el endpoint `/next-exercise`.

```python
# Estado inicial del estudiante
student_state = {
    'student_id': 1,
    'current_level': '1a',
    'score': 0
}

# Llamada al backend
response = requests.post('http://127.0.0.1:8000/next-exercise', json=student_state)
print('Respuesta del backend:', response.json())
```

## 2. Visualizar la red de decisiones (dummy)
Aquí creamos una red de decisiones de ejemplo. Puedes reemplazarla por la real cuando la tengas.

```python
# Creamos una red de decisiones dummy
G = nx.DiGraph()
levels = ['1a', '1b', '2a', '2b', '2c', '3a', '3b', '3c']
for i in range(len(levels)-1):
    G.add_edge(levels[i], levels[i+1], action=f'go_{levels[i+1]}')

# Visualización con networkx
plt.figure(figsize=(8,4))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=14, arrows=True)
edge_labels = nx.get_edge_attributes(G, 'action')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title('Red de Decisiones (Dummy)')
plt.show()
```

## 3. Graficar el progreso de un estudiante
Simulamos el avance del estudiante y graficamos su progreso.

```python
# Simulación de progreso
progreso = []
niveles = ['1a', '1b', '2a', '2b', '2c', '3a', '3b', '3c']
score = 0
for nivel in niveles:
    score += random.randint(5, 15)
    progreso.append({'nivel': nivel, 'score': score})

# Graficar progreso
plt.figure(figsize=(8,4))
plt.plot([p['nivel'] for p in progreso], [p['score'] for p in progreso], marker='o')
plt.xlabel('Nivel')
plt.ylabel('Score acumulado')
plt.title('Progreso del Estudiante (Simulado)')
plt.grid(True)
plt.show()
```
